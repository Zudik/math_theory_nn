```python
import plotly.graph_objects as go
import numpy as np
```

```python
def threshold_func(x):
    return 1 if x >= 0 else 0
```

```python
assert threshold_func(1) == 1
assert threshold_func(0) == 1
assert threshold_func(-1) == 0
```

```python
x = np.linspace(-10, 10, 1000)
y = [threshold_func(i) for i in x]
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=x,
        y=y
    )
)
fig.update_layout(
    title='Treshold function',
    autosize=False,
    width=800,
    height=500,
    yaxis={'title_text':'Y'},
    xaxis={'title_text':'X'}
)
fig.show()
```
